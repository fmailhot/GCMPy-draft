# Using Hood - Hud continuum data from Johnson 1990 (J Phon)

```python
%load_ext autoreload
import GCMPy as gp
%aimport GCMPy
%autoreload 1
import math
import random
import matplotlib.pyplot as plt
#%matplotlib inline
import matplotlib.ticker as ticker
import numpy as np
import pandas as pd
from pandas import DataFrame
from scipy.optimize import minimize
import seaborn as sns

sns.set(style='ticks', context='notebook')
colors=["#045447","#A6262E","#FFBB00","#253060","#8EAB27","#36B5AA"]
sns.set_palette(colors)
```

```python
pb52=pd.read_csv('pb52.csv')
pbbark = gp.HzToBark(pb52,["F0","F1","F2","F3"])
pbbark.sample(5)
```

```python
j90=pd.read_csv('johnson1990.csv')
j90bark=gp.HzToBark(j90,["F0","F1","F2","F3"])
j90bark
```

```python
cval=25
dimsvals={'z0':1,'z1':.761,'z2':.681,'z3':.407}
dimslist = list(dimsvals.keys())
catslist=['vowel','type']
```

```python
startvowel = "FOOT"
endvowel = "STRUT"
speakerlist = [120,240]
sptype = "F0"
stimdata = j90bark.copy()
clouddata = pbbark.copy()
datalist = []


for sp in speakerlist:
    # set start and end
    strt = {"vowel":startvowel,sptype:sp}
    nd = {"vowel":endvowel,sptype:sp}
    # make continuum
    cnt = gp.continuum(stimdata,strt,nd,dimslist,steps=7,stimdetails=False)
    # make choices
    chcs = gp.multicat(cnt,clouddata,catslist,dimsvals,cval,exclude_self=True, N=1,fc={"vowel":[startvowel,endvowel]})
    datalist.append(chcs)

gp.cpplot(datalist,"vowel", speakerlist)

```

```python
# make the 240 continuum
start = {"vowel":"FOOT","F0":240}
end = {"vowel": "STRUT","F0":240}
dimensions = ["F0","F1","F2","F3","z0","z1","z2","z3"]
cont240 = gp.continuum(j90bark,start,end,dimensions,steps=7,stimdetails=False)
cont240
```

### Every stimulus is categorized as w, FOOT

```python
choices240 = gp.multicat(cont240,pbbark,catslist,dimsvals,cval,exclude_self=True, N=1,fc={"vowel":["FOOT","STRUT"]})
```

```python
choices240
```

### If we use Hz not Bark, probability for FOOT = 1.0

```python
dimsvals_240={'F0':1,'F1':.761,'F2':.681,'F3':.407}
dimslist_240 = list(dimsvals_240.keys())

gp.multicat(cont240,pbbark,catslist,dimsvals_240,cval,exclude_self=True, N=1,fc={"vowel":["FOOT","STRUT"]})
```

### If we set w = 1 for z0-z2, .5 for z3, very similar to first result

```python
dimsvals_240={'z0':1,'z1':1,'z2':1,'z3':0.5}
dimslist_240 = list(dimsvals_240.keys())

gp.multicat(cont240,pbbark,catslist,dimsvals_240,cval,exclude_self=True, N=1,fc={"vowel":["FOOT","STRUT"]})
```

### If we take away the Forced Choice aspect, it's still all FOOT

```python
dimsvals_240=dimsvals
dimslist_240 = list(dimsvals_240.keys())
gp.multicat(cont240,pbbark,catslist,dimsvals_240,cval,exclude_self=True, N=1,fc=None)
```

### If we try adding priming for type, it still really wants to pick FOOT

```python
dimsvals_240=dimsvals
dimslist_240 = list(dimsvals_240.keys())

bias_on="type"
terms = {"vowel":["FOOT","STRUT"]}

for i in [2.5, 5, 7.5, 10, 25, 50, 75, 100,250,500,1000,10000,100000]:
    biasval=i
    
    mbias = gp.multicatprime(cont240,pbbark,catslist,dimsvals_240,cval,cat=bias_on,catbias={"w":1,"c":1, "m": biasval},exclude_self=True,fc=terms)
    wbias = gp.multicatprime(cont240,pbbark,catslist,dimsvals_240,cval,cat=bias_on,catbias={"c":1,"m":1, "w": biasval},exclude_self=True,fc=terms)
    nobias = gp.multicatprime(cont240,pbbark,catslist,dimsvals_240,cval,cat=bias_on,catbias={"c":1,"m":1, "w": 1},exclude_self=True,fc=terms)
    
    print(i)
    gp.cpplot([nobias,mbias,wbias],"vowel",["No bias","m","w"])

    
```

### If you increase N for FOOT and STRUT, you still get FOOT

```python
dimsvals_240=dimsvals
dimslist_240 = list(dimsvals_240.keys())

bias_on="vowel"
nb = {"FOOT":1,"STRUT":1, "FLEECE":1, "KIT":1, "DRESS":1, "TRAP":1,"PALM":1,"THOUGHT":1, "GOOSE":1, "NURSE":1}

terms = {"vowel":["FOOT","STRUT"]}

for i in [2.5, 5, 7.5, 10, 25, 50, 75, 100,250,500,1000,10000,100000]:
    biasval=i
    cat_bias = {"FOOT":i,"STRUT":i, "FLEECE":1, "KIT":1, "DRESS":1, "TRAP":1,"PALM":1,"THOUGHT":1, "GOOSE":1, "NURSE":1}
   
    nobias = gp.multicatprime(cont240,pbbark,catslist,dimsvals_240,cval,cat=bias_on,catbias=nb,exclude_self=True,fc=terms)
    fcbias = gp.multicatprime(cont240,pbbark,catslist,dimsvals_240,cval,cat=bias_on,catbias=cat_bias,exclude_self=True,fc=terms)
    
    print(i)
    gp.cpplot([nobias,fcbias],"vowel",["No bias","FC bias"])

    
```
